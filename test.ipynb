```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
test = pd.read_csv('./test.csv')
```

```python
xp = pd.read_csv('./test.csv')['xp']
yp = pd.read_csv('./test.csv')['yp']
fp = pd.read_csv('./test.csv')['fp']
```

```python
q = np.array([xp,yp,fp])
```

```python
def drawRobot(x,y,q,s,h):
    p=np.zeros(36).reshape(12,3)
    p[0,:]=[1,1/7,1/s]
    p[1,:]=[-3/7,1,1/s]
    p[2,:]=[-5/7,6/7,1/s]
    p[3,:]=[-5/7,5/7,1/s]
    p[4,:]=[-3/7,2/7,1/s]
    p[5,:]=[-3/7,0,1/s]
    p[6,:]=[-3/7,-2/7,1/s]
    p[7,:]=[-5/7,-5/7,1/s]
    p[8,:]=[-5/7,-6/7,1/s]
    p[9,:]=[-3/7,-1,1/s]
    p[10,:]=[1,-1/7,1/s]
    p[11,:]=[1,1/7,1/s]
    #
    p=s*p*100
    #
    r=np.zeros(6).reshape(3,2)
    r[0,:]=[np.cos(q),np.sin(q)]
    r[1,:]=[-np.sin(q),np.cos(q)]
    r[2,:]=[x,y]
    #
    p=np.dot(p,r)
    X=p[:,0]
    Y=p[:,1]
    h.plot(X,Y,'r-')
```

```python
fig,ax=plt.subplots()
ax.axis('equal')
ax.plot(xp,yp,color='blue',linestyle='dashed',linewidth=1)
plt.grid()
plt.title("Top view: robot trajectory")
plt.xlabel("x, m")
plt.ylabel("y, m")
plt.show(block=False)
#
for i in range(0,len(xp)-1,int(round(len(xp)/20))):
    drawRobot(xp[i],yp[i],fp[i],0.01,ax)
#
plt.show()
```
